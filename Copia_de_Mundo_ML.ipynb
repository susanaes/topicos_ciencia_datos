```
# https://geopandas.org/gallery/plotting_basemap_background.html#sphx-glr-gallery-plotting-basemap-background-py
# usando un fondo de un mapa real
 
import contextily as ctx
```

```
# https://epsg.io/3857 - la base (tile) estandar con el sistema de cooredenadas que ocupa google maps etc
Final_df_epsg_3857 = Final_df.to_crs(epsg=3857)
ax = Final_df_epsg_3857.plot(figsize=(20, 10), alpha=0.8, legend=True, cmap="OrRd", edgecolor='k', column='RESULTADO')
ctx.add_basemap(ax)
```
